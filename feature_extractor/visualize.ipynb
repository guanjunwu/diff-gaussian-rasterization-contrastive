```python
import numpy as np
from sklearn.decomposition import PCA
import torch
import matplotlib.pyplot as plt

features = torch.load("../data/dynerf/cut_roasted_beef/cam05/features.pt")
emb = features["0000.png"]
print(emb.shape)
emb = torch.nn.functional.interpolate(emb[None, ...], size=(1014, 1352), mode='bilinear').squeeze(0)
emb = torch.nn.functional.normalize(emb, dim=0).permute(1,2,0)
# print(emb.shape)
# C, H, W = emb.shape
X = emb.flatten(0, 1).cpu().numpy()
np.random.seed(6)
pca = PCA(n_components=3)
pca.fit(X)
X_rgb = pca.transform(X).reshape(*emb.shape[:2], 3)
plt.imshow(X_rgb)
plt.axis("off")
plt.show()
```
